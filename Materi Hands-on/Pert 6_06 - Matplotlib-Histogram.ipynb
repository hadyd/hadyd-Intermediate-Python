```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

If we do survey to some people, it might be able for us to plot it in bar graph. But what if we have 100 or more people? That's why we need histogram. \
Histogram allows us to create bins for our data

```python
plt.hist(ages)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

```python
plt.hist(ages, bins=5)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

add edge color

```python
plt.hist(ages, bins=10, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

We can pass our own list of value (bins). It allows us to have control

```python
#plot 10yrs difference
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 15, 20, 30, 40, 50, 60]
```

```python
plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

We can exclude a data

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [20, 30, 40, 50, 60]

plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

### Let's use real world data

```python
data = pd.read_csv('data_histo.csv')
ids = data['Responder_id']
ages = data['Age']

bins = [10, 20, 30, 40, 50, 60, 80, 90, 100]
```

```python
plt.hist(ages, bins=bins, edgecolor='k')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

```python
plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```

Add median line

```python
median_age = 29
color = 'red'

plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.axvline(median_age, color=color, label='Age Median')

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')
plt.legend()

plt.show()
```

```python
median_age = 29
color = 'red'

plt.hist(ages, bins=bins, edgecolor='k', log=True)

plt.axvline(median_age, color=color, label='Age Media', linewidth=2)

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.show()
```
